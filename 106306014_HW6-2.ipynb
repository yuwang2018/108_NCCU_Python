### 計算班級隨機成績平均、標準差及分布圖

```python
%matplotlib inline 

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib as mpl
```

```python
mpl.rcParams.keys()
# https://blog.csdn.net/m0_37362454/article/details/82180762
# https://blog.csdn.net/ztf312/article/details/49933497
```

```python
grade = np.random.randint(101,size = 60)
glist = list(grade)
print(glist)
```

```python
def passg(x):
    if x >= 60 :
        return True
    else:
        return False
```

```python
passpeople = 0
notpass = 0
for x in glist:
    if passg(x) == True:
        passpeople = passpeople + 1
    else :
        notpass = notpass + 1

avg = np.mean(grade)
std = np.std(grade)
num = np.linspace(0,60,60)
print(f"全班及格人數為{passpeople}人,不及格人數為{notpass}人\n")
print(f'平均分數為{avg}分')
print(f'標準差為{std}')
print("以下是成績分布圖")
mpl.rcParams['xtick.labelsize'] = 16
plt.rcParams["ytick.color"] = 'b'
plt.plot(grade,num,'m^')
```
