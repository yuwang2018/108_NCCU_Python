# 地震次數統計

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('https://quality.data.gov.tw/dq_download_csv.php?nid=12730&md5_url=ab16ef0a4f6bea0bd1b70dabcc516baf')
df
```

### Result:歷年來發生的地震總數

```python
df1 = df.groupby(by='年份').sum()
df1
```

### 折線圖

```python
df1.plot()
```

### 直方圖

```python
df1.hist(bins = 20, color='red')
```

### 篩出地震規模為6≦M<7的資料

```python
filtter = (df['規模'].str.contains("6≦M<7"))
df2 = df[filtter]
df2
```

### Result:歷年來發生地規模為6≦M<7的地震的次數趨勢圖

```python
df2.plot('地震個數',color='orange')
```
