# 全國大學學生數分析

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('https://quality.data.gov.tw/dq_download_csv.php?nid=6231&md5_url=c3e78c01d880aa9e9e65ca95d444d945')
```

```python
df
```

```python
df1 =df[["學校名稱","等級別","一年級男生","一年級女生","縣市名稱"]]
df1
```

### 查詢大學學士班一年級男女人數

```python
df2 = df1[(df1.等級別 == 'B 學士')]
df2
```

```python
df2.head()
```

### 依縣市名稱排序

```python
df3 = df2.sort_values(by=['縣市名稱'],ascending = True)
df3
```

### 查詢全台北市大學學士班一年級男女人數

```python
df4 = df3[(df3.縣市名稱 == '30 臺北市')]
df4
```

### 台北市各大學一年級男生人數與一年級女生人數之關係圖

```python
plt.scatter(df4.一年級男生,df4.一年級女生)
```
