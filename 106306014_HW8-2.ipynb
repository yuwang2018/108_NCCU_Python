# 全國大學學生數分析

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('https://quality.data.gov.tw/dq_download_csv.php?nid=6231&md5_url=c3e78c01d880aa9e9e65ca95d444d945')
```

```python
df
```

```python
df1 =df[["學校名稱","等級別","一年級男生","一年級女生","縣市名稱"]]
df1
```

### 查詢大學學士班一年級男女人數

```python
df2 = df1[(df1.等級別 == 'B 學士')]
df2
```

```python
df2.head()
```

### 查詢一年級男(先)/女(後)生人數由高至低排序

#### @排序結果怪怪的@

```python
df2.sort_values(by=['一年級男生','一年級女生'],ascending = False)
```

### 一年級男生人數與一年級女生人數之關係圖

```python
plt.scatter(df2.一年級男生,df2.一年級女生)
```
