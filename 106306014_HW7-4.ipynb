### 畫出你想要的函數圖型

```python
%matplotlib inline 

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
msg1 = input("希望x軸的起始值為(數字):\n")
msg2 = input("希望x軸的終值為(數字):\n")
msg3 = input("希望x軸有多少個點(整數):\n")
msg4 = input("希望y等於 1.sin(x) 2.cos(x) 3.tan(x) 的哪一種呢？(請輸入數字):\n")
n1 = float(msg1)
n2 = float(msg2)
n3 = int(msg3)
n4 = float(msg4)

x = np.linspace( n1, n2 ,n3)
if 1 == n4:
    y = np.sin(x)
    plt.scatter(x[y<0] ,y[y<0], c='#FFC51B')
    plt.plot(x,y)
elif 2 == n4:
    y = np.cos(x)
    plt.scatter(x[y<0] ,y[y<0], c='#FFC51B')
    plt.plot(x,y)
else:
    y = np.tan(x)
    plt.scatter(x[y<0] ,y[y<0], c='#FFC51B')
    plt.plot(x,y)
```
