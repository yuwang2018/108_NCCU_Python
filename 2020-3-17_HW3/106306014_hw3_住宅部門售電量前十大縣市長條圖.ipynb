```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML
import numpy as np
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei']
```

```python
#data:https://data.gov.tw/dataset/29935
t1 = pd.read_csv('https://quality.data.gov.tw/dq_download_csv.php?nid=29935&md5_url=01410d8f98b8e14419aab3ad9802e371')
```

```python
t1
```

```python
t2 = t1[(t1.日期 == '2020年02月')]
t2
```

```python
t3 =t2[['日期','縣市',"住宅部門售電量(度)"]]
t3 = t3.drop(22)
```

```python
basic = t3[t3['日期']=='2020年02月'].sort_values('住宅部門售電量(度)').tail(10)  # 將資料鎖定在'2001-01-01' 並以人數排序取前十名
fig, ax = plt.subplots(figsize=(15, 8))  
ax.barh(basic['縣市'], basic['住宅部門售電量(度)'])
```

```python
colors = dict(zip(
    ['新竹縣',"雲林縣", "屏東縣", "彰化縣", "台南市", "高雄市", "台北市", "桃園市","台中市","新北市"],
    ["#a18cd1", "#fbc2eb", "#fad0c4", "#d4fc79", "#8fd3f4", "#fad0c4", "#ff9a9e","#f6d365", "#fbc2eb", "#a6c1ee"]
))
#色票網站：https://webgradients.com/
```

```python
fig, ax = plt.subplots(figsize=(16, 9))

def race_barchart(input_year):
    dff = t3[t3['日期'].eq(input_year)].sort_values(by='住宅部門售電量(度)', ascending=True).tail(10)
    ax.clear()

    ax.barh(dff['縣市'], dff['住宅部門售電量(度)'], color=[colors[x] for x in dff['縣市']],height=0.8)
    dx = dff['住宅部門售電量(度)'].max() / 200
    
    for i, (value, name) in enumerate(zip(dff['住宅部門售電量(度)'], dff['縣市'])):
        ax.text(0, i,name+' ',size=16, weight=600, ha='right', va='center') #將每個國家名稱放到對應的長條旁邊
        ax.text(value+dx, i,f'{value:,.0f}',  size=16, ha='left',  va='center') 
            
    ax.text(1.0, 0.2, input_year[:7].replace('-','/'), transform=ax.transAxes, color='#777777', size=72, ha='right', weight=1000) #增加日期
    ax.text(0, 1.06, '售電量(度)', transform=ax.transAxes, size=14, color='#777777') #小標題
    ax.text(0.67, 0.14, '總售電量(度):'+str(int(dff['住宅部門售電量(度)'].sum())), transform=ax.transAxes, size=24, color='#000000',ha='left') #計算總度數
    ax.tick_params(axis='x', colors='#777777', labelsize=12) #調整橫軸顏色、標籤大小
    ax.xaxis.set_ticks_position('top') #將橫軸放到頂端
    ax.set_yticks([])
    ax.margins(0, 0.01)
    ax.grid(which='major', axis='x', linestyle='-') #增加格線
    ax.text(0, 1.15, '住宅部門售電量前十大縣市(度)',
                transform=ax.transAxes, size=24, weight=600, ha='left', va='top') #標題

    plt.box(False)
    
race_barchart('2020年02月')
```
