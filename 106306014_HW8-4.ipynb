# 各縣市年度職災案例分析

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('https://quality.data.gov.tw/dq_download_csv.php?nid=6644&md5_url=abad3839bcab316d2bd9dec57e713174')
```

```python
df
```

```python
#https://chrisalbon.com/python/data_wrangling/pandas_dropping_column_and_rows/
df1=df.drop(df.index[0])
```

```python
df1.columns=['縣市別','行業別','媒介物','災害類型','發生日期','死亡人數','受傷人數']
```

```python
df1
```

### 各縣市職災傷亡人數統計

```python
#https://medium.com/jameslearningnote/%E8%B3%87%E6%96%99%E5%88%86%E6%9E%90-%E6%A9%9F%E5%99%A8%E5%AD%B8%E7%BF%92-%E7%AC%AC2-3%E8%AC%9B-pandas-%E5%9F%BA%E6%9C%ACfunction%E4%BB%8B%E7%B4%B9-series-dataframe-selection-grouping-447a3fa90b60
df2 = df1.groupby(by='縣市別').sum()
df2
```

### 加入總傷亡人數

```python
df2["總傷亡人數"]=df2.loc[:,"死亡人數":"受傷人數"].sum(axis=1)
df2
```

### 各縣市之總傷亡人數排序(高至低)

```python
df2.sort_values(by='總傷亡人數',ascending = False)
```

### 各縣市之死亡人數、受傷人數、總傷亡人數之平均值

```python
df2.groupby(by='縣市別').mean()
```
