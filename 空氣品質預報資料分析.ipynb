```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## 讀入資料

```python
df1 = pd.read_csv('https://opendata.epa.gov.tw/ws/Data/AQFN/?$format=csv')
```

```python
df1
```

## 資料清理

```python
# 把Nan變成0
df2 = df1.fillna(0) 
df2
```

## 刪除不必要欄位

```python
coln = ['MinorPollutant', 'MinorPollutantAQI']
df3 = df2.drop(coln, axis=1)
df3
```

## 將欄位名稱改為中文

```python
df3.columns=['預報內容','空品區','指標污染物','空氣品質指標預報值','預報日期','發布時間']
df3
```

## 排序資料

```python

df4 = df3.sort_values(by="空氣品質指標預報值", ascending=False)
df4
```

## 將欄位反向

```python

df5= df4.loc[:3, ::-1]
df5
```

## 將aqi依數值大小改變顏色

```python
def color(value):
    if type(value) == int:
        if value <=50:
            color ='green'
            return ("color:%s" %color)
        elif value <= 100:
            color='yellow'
            return ("color:%s" %color)
        else:
            color ='darkblue'
            return ("color:%s" %color)
    else:
        color ='darkblue'
        return ("color:%s" %color)
```

```python
df5.style.applymap(color)
```

## 顯示aqi數值分布

```python
plt.scatter(df3.index,df3.空氣品質指標預報值)
```

### tqdm

```python
from tqdm import tqdm_notebook
tqdm_notebook().pandas()
#clear_output()

# 只需將 `apply` 替換成 `progress_apply`
df5['空氣品質指標預報值'] = df5.空氣品質指標預報值.progress_apply(lambda x: '50' if x else '100')
```
