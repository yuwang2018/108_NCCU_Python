## 來臺旅客人數分析(亞洲地區)

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
table = pd.read_html('http://stat.motc.gov.tw/mocdb/stmain.jsp?sys=220&ym=9000&ymt=10700&kind=21&type=1&funid=b710101&cycle=4&outmode=0&compmode=0&outkind=1&fldspc=1,11,&rdm=immlliqn')
```

## 調整表頭名稱

```python
t1.columns=['年','亞洲地區','港澳','中國','日本','南韓','馬來西亞']
```

### 各數值變數的描述統計

```python
t1.describe()
```

### 將亞洲地區來台旅遊人數作排序，得知哪幾年的旅遊人數較多

```python

t1.sort_values(by='亞洲地區',ascending = False)
```

### 歷年來亞洲地區來台旅遊人數成長趨勢

```python

t1.亞洲地區.plot()
```
