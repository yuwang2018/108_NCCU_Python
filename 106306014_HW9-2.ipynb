# 全國四大超商營業資料分析

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('https://quality.data.gov.tw/dq_download_csv.php?nid=32086&md5_url=7339a804f8a2ddca126bfc7abdeb2a0e')
```

```python
#分公司狀態:01核准設立，02停業，03撤銷/廢止
df
```

### 取還在營業的公司資料

```python
filtter = (df['分公司狀態'] == 1)
df0 = df[filtter]
df0
```

### 算每家超商的營業分公司數

```python
df1 = df0.groupby(by='公司名稱').sum()
df1
```

### 排序超商的營業分公司數

```python
df1.sort_values(by="分公司狀態",ascending = False)
```
