### 會回應的成績計算機

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
while True:
    spam = input("b(ゝ∀･)b ：請輸入您預計要算的成績。\n")
    num = spam.split(' ')
    def spam(g):
        return g>= 60
    
    glist = list(num)
    num = len(glist)
    
    up = 0
    down = 0
    gsum = 0
    i = 0
    for i in range(num):
        gsum = gsum + int(glist[i]) 
        i = i + 1
        
    for grade in glist:
        grade = float(grade)
        spam(grade)
        if (spam(grade) == True):
            up = up + 1
        else:
            down = down + 1
            
    avg = gsum / num
    print(f"平均成績為 {avg}")
    print(f"及格人數/次數為 {up}")
    print(f"不及格人數/次數為 {down}")
    
    a = up/(up+down)*100
    if (a > 80):
        print(f"及格人數/次數佔全體的 {str(a)} % ,表現超棒的啦！送你一首棒棒歌～")
        video1 = YouTubeVideo('yO6nzuY6WJQ')
        display(video1)
    elif(a > 60):
        print(f"及格人數/次數佔全體的 {str(a)} % ,表現不錯～繼續保持下去^^")
        picture1 = Image('哎呦不錯喔.jpg')
        display(picture1)
    elif(a > 40):
        print(f"及格人數/次數佔全體的 {str(a)} % ,嗯...感覺還沒發揮真正的實力ㄟ，多多努力好嗎><")
        video2 = YouTubeVideo('T-VHLCxd2_s')
        display(video2)
    else:
        print(f"及格人數/次數佔全體的 {str(a)} % ,請相信人有無限的潛能，不要輕易放棄！")
        picture2 = Image('勵志語錄圖.jpg')
        display(picture2)
    
```
