# 比大小遊戲
### 計算隨機和電腦比賽的輸贏次數

```python
money = input('請輸入本金:\n')
m = int(money)
w = 0
n = 0
l = 0
while (m>=0):
       
    #去除J、Q、和K
    x1 = randint(1, 6)#user
    x2 = randint(1, 6)#computer
        
    if (x1 > x2):
        
        w = w+1
        print('You win!\n')
        m = 100 + m
        print('目前擁有的本金為:',m)
        
    elif (x1 == x2):
        
        n = n+ 1 
        print('You get one more chance!')
            
    else :
        
        l = l+1
        print('You lose!\n')
        m = m - 100
        print('目前擁有的本金為:',m)
        
        
    if(m == 0):
        print("共贏了",w,'次，輸了',l,'次，平手',n,'次')
        break
        
```
