# 使用contour畫圖

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = np.array([50,55,60,65,70,75])
y = np.array([10,15,20,25,30,35])
X, Y = np.meshgrid(x, y)
```

```python
K = np.random.randint(1,200,(6,6))
K
```

```python
plt.contour(X,Y,K)
plt.scatter(X.ravel(), Y.ravel(),c=K.ravel())
```

```python
plt.contourf(X,Y,K)
```
