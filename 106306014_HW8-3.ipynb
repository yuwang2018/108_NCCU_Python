## 半導體股市當日行情分析

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_html('https://tw.stock.yahoo.com/s/list.php?c=%A5b%BE%C9%C5%E9&rr=0.52787500%201573544053',encoding='big5')
```

```python
len(df)
```

```python
stock = df[2]
```

```python
stock
```

```python
stock.drop('選擇',axis = 1, inplace = True)
stock
```

### 跌幅最高前五家公司

```python
stock.sort_values(by='漲跌',ascending = False).head(5)
```

```python
#取絕對值
stock["買進與賣出之差值"]= abs(stock.買進 - stock.賣出)
```
