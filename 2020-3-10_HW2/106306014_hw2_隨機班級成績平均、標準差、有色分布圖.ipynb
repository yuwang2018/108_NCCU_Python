```python
%matplotlib inline 

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib as mpl
```

```python
mpl.rcParams.keys()
```

```python
grade = np.random.randint(101,size = 80)
glist = list(grade)
print(glist)
```

```python
def goodgrade(x):
    if x >= 60 :
        return True
    else:
        return False
```

```python
passp = 0
notpass = 0
for x in glist:
    if goodgrade(x) == True:
        passp = passp + 1
    else :
        notpass = notpass + 1

avg = np.mean(grade)
std = np.std(grade)
num = np.linspace(0,80,80)
print(f"全班及格人數為{passp}人,不及格人數為{notpass}人\n")
print(f'平均分數為{avg}分')
print(f'標準差為{std}')
print("以下是成績分布圖")
mpl.rcParams['xtick.labelsize'] = 16
plt.rcParams["ytick.color"] = 'm'
plt.plot(grade,num,'c^')
```
